```python
# importing all the required python libraries
import numpy as np
from scipy.stats import beta
from matplotlib import pyplot as plt
import seaborn
plt.style.use('ggplot')
import matplotlib.animation as animation
import random
# plt.style.available
```

```python
#generating a random binary array where 1 corresponds to heads and 0 to tails
arr = [1 if i < 48 else 0 for i in range(160) ]
random.shuffle(arr)
print(arr)
```

```python
# Initializing Parameters for Beta distribution
a=6
b=9
```

```python
# Plotting prior distribution using beta distribution with a=6 and b=9
def plot_beta(a, b, m = 0, l = 0):
    x = np.linspace(0, 1, 100)
    y = beta.pdf(x, a + m, b + l)
    plt.plot(x, y)
    plt.show()
    
plot_beta(a, b)
# plt.savefig('Beta_dist.png')
```
